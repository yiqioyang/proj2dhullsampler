```python
var_path = "/glade/work/qingyuany/repo_data/spatialtuning/emulated_ensemble/"

mean_paths = glob.glob(var_path + "*mean*", recursive=True)
```

```python
ppe_pd = pd.read_csv("/glade/work/qingyuany/repo_data/spatialtuning/ppe_zonal_v4_iteration1.csv", index_col=0)
obs_pd = pd.read_csv("/glade/work/qingyuany/repo_data/spatialtuning/obs_zonal_v4_iteration1.csv", index_col=0)
```

```python
threshold_level = 2.0
```

```python
tf_masks = []

for path in tqdm(mean_paths):
    var_name_file = path.split("/")[-1].split("_mean_std_")[1]
    if "zonal" in var_name_file:
        cli_temp = var_name_file.split("_zonal")[0]
        
    else:
        cli_temp = var_name_file.split("_")[0]  #% xxx

    print(cli_temp)
    
    
    var_name = var_name_file.split(".")[0]
    
    emulated_mean_std = pd.read_csv(path,index_col=0)
    emulated_mean = emulated_mean_std.iloc[:,0]
    emulated_std = emulated_mean_std.iloc[:,1]
    #emulated_std = pd.read_csv(var_path + "gp_std_" + var_name_file, index_col=0) 
    
    
    obs_temp = obs_pd.loc[var_name].values[0]
    y_ppe = ppe_pd[var_name]

    yscale = y_ppe.std()
    ymu = y_ppe.mean()            
    emulated_mean = emulated_mean * yscale + ymu
    emulated_std = emulated_std * yscale

    
    
    if var_name.split("_")[0] == "PRECT":
        emulated_mean = emulated_mean * 1000 * 3600 * 24
        emulated_std = emulated_std * 1000 * 3600 * 24

    
    temp_tf_mask = ((emulated_mean - threshold_level * emulated_std) < obs_temp) & ((emulated_mean + threshold_level * emulated_std) > obs_temp)
    temp_tf_mask.name = var_name
    tf_masks.append(temp_tf_mask)

```
